### Convert some object columns into numerical type or float columns

```python
columns = ['Production Budget', 'Opening Weekend', 'Domestic Box Office', 'Worldwide Box Office']

for col in columns:
    mcu_boxoffice_clean.loc[:,col] = mcu_boxoffice_clean[col].str.replace('[$,]','', regex = True)
    mcu_boxoffice_clean.loc[:,col] = pd.to_numeric(mcu_boxoffice_clean[col], errors = 'coerce')
    mcu_boxoffice_clean.loc[:,col] = mcu_boxoffice_clean[col]/1000000

mcu_boxoffice_clean.head()
```

### Find out how many times the Domestic box office is of Opening weekend

```python
mcu_boxoffice_clean["domestic opening ratio"] = mcu_boxoffice_clean["Domestic Box Office"] / mcu_boxoffice_clean["Opening Weekend"]
mcu_boxoffice_clean.head()
```

```python
highest_domestic_opening = mcu_boxoffice_clean.loc[mcu_boxoffice_clean['domestic opening ratio'].idxmax()]
highest_domestic_opening
```

### Find out how many times the Domestic box office is of Worldwide Box office

```python
mcu_boxoffice_clean["domestic worldwide ratio"] = mcu_boxoffice_clean["Worldwide Box Office"] / mcu_boxoffice_clean["Domestic Box Office"]
mcu_boxoffice_clean.head()
```

### Create new columns to donate domestic and overseas percentage share in box office

```python
mcu_boxoffice_clean["domestic % share"] = (mcu_boxoffice_clean["Domestic Box Office"]/mcu_boxoffice_clean["Worldwide Box Office"]) * 100
mcu_boxoffice_clean["overseas % share"] = 100 - mcu_boxoffice_clean["domestic % share"]
mcu_boxoffice_clean.head()
```

## Aggregation and Grouping

### What is mean box office by Opening , Domestic and Worldwide?

```python
mean_boxoffice = mcu_boxoffice_clean[['Opening Weekend','Domestic Box Office','Worldwide Box Office']].mean()
mean_boxoffice
```

### Create a new column by series - Iron Man 1, Iron Man 2, Iron Man 3 - Series Iron Man

```python
series_keywords = ['Iron Man', 'Hulk', 'Thor', 'Captain America', 'Marvel', 'Black Panther', 'Spider-Man',
         'Doctor Strange', 'Eternals', 'Ant-Man', 'Guardians of the Galaxy', 'Black Widow', 'Shang-Chi',
         'Avengers']                           

def series(title):
    for keyword in series_keywords:
        if keyword in title:
            return keyword
    return 'other'

mcu_boxoffice_clean['series'] = mcu_boxoffice_clean['Title'].apply(series)     #apply function
mcu_boxoffice_clean.head()
```

### Calculate the frequency of movies in a series

```python
mcu_boxoffice_clean['series'].value_counts()   
```

### Calculate total domestic and worldwide boxoffice by series

```python
total_by_series = mcu_boxoffice_clean.groupby('series')[['Domestic Box Office','Worldwide Box Office']].sum()
total_by_series
```

### Calculate the mean opening deomestic and worldwide box office of every series

```python
mean_by_series = mcu_boxoffice_clean.groupby('series')[['Opening Weekend','Domestic Box Office','Worldwide Box Office']].mean()
mean_by_series
```

### Calculate the min , max and mean of every series

```python
stats = mcu_boxoffice_clean.groupby('series')[['Opening Weekend','Domestic Box Office','Worldwide Box Office']].agg(['min','max','mean'])
stats
```

## Filtering

### Filter the data based on movies which has more domestic boxoffice than their production budget

```python
filtered_data = mcu_boxoffice_clean[mcu_boxoffice_clean['Domestic Box Office'] > mcu_boxoffice_clean['Production Budget']]
filtered_data
```

### Transform the data for all series by converting the figues into billion

```python
mcu_transformed = mcu_boxoffice_clean.copy()
for col in ['Opening Weekend', 'Domestic Box Office', 'Worldwide Box Office']:
    mcu_transformed[col] = mcu_boxoffice_clean.groupby('series')[col].transform(lambda x: x/1000)
    
mcu_transformed 
```

### Use apply function to sum all Worldwide box office and domestion box office figues by groupby the series


```python
sum_by_series = mcu_boxoffice_clean.groupby('series')[['Domestic Box Office','Worldwide Box Office']].apply(lambda x: x.sum())
sum_by_series
```

## Groupby
